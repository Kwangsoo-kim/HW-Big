```python
import numpy as np # csv파일을 numpy배열로 읽기
import tensorflow.keras.utils as utils # 라벨링
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation
```

## 1.1 데이터 셋 생성

```python
dataset = np.loadtxt('data/pima-indians-diabetes.csv',delimiter=',')
dataset.shape
```

```python
# 학습용 데이터(훈련데이터)
x_train = dataset[:700,:8]
y_train = dataset[:700,8]
# 테스트 데이터
x_test = dataset[700:,:8]
y_test = dataset[700:,8]
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

## 1.2 모델 구성하기

```python
model = Sequential()
model.add(Dense(64,input_dim=8,activation='relu'))# input_dim = 8 최종출력 1
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='sigmoid')) # 원 핫 인코딩 안했을땐 sigboid
```

## 1.3 학습과정 설정

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

## 1.4 학습시키기

```python
hist = model.fit(x_train,y_train,epochs=1500,batch_size=32)
```

## 1.5 모델평가하기

```python
score = model.evaluate(x_test,y_test)
```

```python
print('loss : %s\naccuracy: %.2f'%(score[0],score[1]))
```
