# 1. XOR

```python
import numpy as np
from tensorflow.keras.models import Sequential # 모델 객체 생성시 필요
from tensorflow.keras.layers import Dense # layer 쌓을 때 필요
import matplotlib.pyplot as plt
```

```python
# 1. traing data set
x_data = [[0,0],
          [0,1],
          [1,0],
          [1,1]]
y_data = [[0],
          [1],
          [1],
          [0]]

# 2. 모델 구성
model = Sequential()
# tensorflow v1 에서 weight & bias에서 layer 생성하는것과 같다
model.add(Dense(10,input_dim=2,activation='relu')) 
model.add(Dense(20,activation='relu')) # 은닉층은 input_dim 쓰지 않는다.
model.add(Dense(10,activation='relu')) # 은닉층은 주로 relu사용 항상은 아니다.
model.add(Dense(1,activation='sigmoid')) # binaryclassification은 sigmoid
model.summary() # 파라미터의 개수 확인

# 3. 모델 학습과정
# tensorflow v1 의 cost가 v2에선 loss
# mse 는 linear regression 할때
# optimizer v1 에서는 learning_rate로 직접 조정 v2에서는 패키지에 제공
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['binary_accuracy'])

# 4. 학습
hist = model.fit(x_data,y_data,epochs=100)
```

```python
# 정확도
for acc in hist.history['binary_accuracy']:
    print(acc,end='\t')
```

```python
# loss 는 그래프로
plt.plot(hist.history['loss'])
plt.show()
```

```python
# 입력 스타일은 "0 0"
while True:
    input_list = input('space로 분리해서 0이나 1을 2개 입력(종료:9)').strip().split()
    input_data = list(map(int,input_list)) #형 변환. map함수는 변수에 한꺼번에 적용시켜준다.
    if input_data[0]==9:
        break;
    print('입력값 : ', input_data)
    print('예측값 : ', int(model.predict([input_data]).round()))
```
