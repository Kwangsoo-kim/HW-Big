**2) subplots()함수로 서브플롯 추가하기**

```python
x = np.arange(0,10,0.01)
x.shape
```

```python
# 방법1.(2행2열 서브플롯)
plt.subplot(2,2,1); plt.plot(x,np.sin(x))
plt.subplot(2,2,2); plt.plot(x,np.cos(x))
plt.subplot(2,2,3); plt.plot(x,np.tanh(x))
plt.subplot(2,2,4); plt.plot(x ,np.sin(x)*np.cos(x))
plt.show()
```

```python
# 방법2 (2행2열 서브플롯)
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(10,6))
axes[0,0].plot(x, np.sin(x),color='r',linewidth=5)
axes[0,0].plot([0,1,2,3],[-1,-0.5,0,0.5],'k:') # 덧그릴 수 있다.
axes[0,1].plot(x,np.cos(x),'c--') # c(청록),--(dashed line)
axes[1,0].plot(x,np.tanh(x),'g')
axes[1,1].plot(x,np.sin(x)*np.cos(x),'r-.')
plt.show()
```

```python
# 방법3 (2행 2열 서브플롯)
x = np.arange(0,7,0.01)
x.shape
```

```python
func_list = [np.sin,np.cos,np.tanh,lambda x : np.sin(x)*np.cos(x)]
fmt_list = ['k:','c--','g','r-.']
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(10,6))
# 2차원 axes를 list(axes.flat)이용해서 중첩for문 안씀
for i, ax in enumerate(axes.flat):
    ax.plot(x, func_list[i](x),fmt_list[i])
```

```python
fig,axes = plt.subplots(ncols=4,figsize=(12,3)) # 1행 4열
for i, ax in enumerate(axes):
    ax.plot(x,func_list[i](x),fmt_list[i])
```

```python
fig,axes = plt.subplots(nrows=4,figsize=(10,14)) #4행 1열
for i , ax in enumerate(axes):
    ax.plot(x,func_list[i](x),fmt_list[i])
```

**4) pyplot.scatter() : 산점도**

## 2.5 다양한 그래프 그리기 예
- 1) fill(),fill_between(),scatter()
- 2) bar(),axvline(),axhline()
- 3) hist(),boxplot(),violinplot()

```python
x = np.linspace(0,10)
y = np.cos(x)
fig,axes = plt.subplots(2,2,figsize=(14,5))
#scatter
axes[0,0].scatter(x,y,s=5,c='r',marker='o')
#fill
axes[0,1].fill(x,y,c='b')
#fill_between
axes[1,1].fill_between(x,y,color='r')
axes[1,1].plot([-1,5],[-1,1],'bo-')
```

```python
fig,axes = plt.subplots(2,2)
axes[0,0].bar([1,2,3],[3,4,5],color='r')
axes[0,0].plot([1,2,3],[3,4,5],color='b')
axes[0,1].bar([0.5,1,2.5],[0,1,2])
axes[0,1].plot([0.5,1,2.5],[0,1,2],'g')
axes[1,0].axvline(0.5)
axes[1,1].axhline(0.75)
```

```python
fig,axes = plt.subplots(2,2)
axes[0,0].hist(iris_df['sepal_length'],bins=20,color='r') # 분포
axes[0,1].hist(iris_df['sepal_length'],bins=20,cumulative=True) # 누적분포
axes[1,0].hist(iris_df['sepal_length'],bins=10,orientation='horizontal') # 옆으로 눕힘
axes[1,1].hist(iris_df['sepal_length'],bins=10,histtype='step')
plt.show()
```

```python
# 겹치기도 가능
fig,axes = plt.subplots(1,2)
axes[0].boxplot(iris_df['sepal_length'])
axes[1].violinplot(iris_df['sepal_length'])
plt.show()
```

## 2.6 그래프 커스터마이징

**1) linestyle, linewidth**

```python
x = np.linspace(0,10,100)
y = np.cos(x)
```

```python
fig,axes = plt.subplots(2,2)
axes[0,0].plot(x,y,linewidth=3,color='r')
axes[0,1].plot(x,y,linestyle='dotted',linewidth=5,color='purple')
axes[1,0].plot(x,y,linestyle='--',linewidth=5,color='b')
axes[1,1].plot(x,y,ls='-.',linewidth=5,color='c') # 요약해서 쓸수있다.
plt.show()
```

**2) text(),annotate()**

```python
fig, axes = plt.subplots(1,2,figsize=(10,5))
axes[0].scatter(x,y,marker='.')
axes[0].text(2,0,'샘플 텍스트',style='italic') #한글 설정해줘야됨
axes[1].scatter(x,y,marker='*',color='r')
axes[1].annotate('Sine',xytext=(2,0.75), # 글자위치
                 xy=(5,0.5),# 화살표가 가리킬 위치
                 arrowprops = dict(arrowstyle='->',
                                   connectionstyle='angle3')               
                ) # 글자가 있을 위치
plt.show()
```

**4) 축과 눈금**

```python
import numpy as np
x = np.linspace(0,10,100)
y = np.cos(x)
fig, axes = plt.subplots(2,2, figsize=(12,5))
# plt.subplots_adjust()을 사용하여 서브플롯간 간격 변경
plt.subplots_adjust(hspace=0.4,wspace=0.3) # hspace는 위아래 여백 wspace 옆여백
plt.suptitle('main title')
axes[0,0].scatter(x,y,marker='.')
axes[0,0].set(title='An Example Axes',ylabel='y-axis',xlabel='x-axis',xlim=[0,12],ylim=[-2,2]) # 타이틀,xy라벨,xy축변경
# 일일히 설정하는 법 set_~~
axes[0,1].scatter(x,y,marker='*',c='b')
axes[0,1].set_xlim(2,6) # x축변경
axes[0,1].set_xlabel('x축 0~5')
axes[0,1].set_ylim(-1.5,1.5)
axes[0,1].set_ylabel('y축 -1.5~1.5')
axes[1,0].scatter(x,y,marker='^',c='c')
# x축 눈금 수정
axes[1,0].set_xticks(range(0,11,3))
axes[1,0].set_xticklabels(['0',3,'change','taht']) # s 없음
axes[1,0].set_yticks([-2,0,1,2])
axes[1,0].set_yticklabels([-2,'개수','는','맞춰서'])
axes[1,1].scatter(x,y,c='r')
# 한단락에 바꾸기
axes[1,1].set(xticks=[0,2,4,6,8,10],xticklabels=['영','이','사','육','팔','십'],title='test')
#격자
axes[1,1].grid(True)
# 그래프 상단 테두리 제거 
axes[1,1].spines['top'].set_visible(False)
# 그래프 하단 테두리 밀어내기 
axes[1,1].spines['bottom'].set_position(('outward',10))
plt.show()
```

**5) 축공유**

```python
x = np.arange(0,10)
y1 = 0.5*x**2
y2 = -1*y1
```

```python
fig, ax1 = plt.subplots(figsize=(10,5))
plt.suptitle('전체 타이틀')
ax1.plot(x,y1,'g^:')
ax1.set(xlabel='x data',ylabel='y1 data')
# twinx : x축 공유하는 그래프
ax2 = ax1.twinx()
ax2.plot(x,y2,'b.--')
ax2.set_ylabel('y2 data', c='b')
# twiny : y축 공유하는 그래프
ax3 = ax1.twiny()
ax3.plot(-x,y1,'ro-')
ax3.set_xlabel('-x data',c='r')
plt.show()
```
